# EDA Automatico 

Generamos un Análisis Exploratorio de Datos de manera automática utilizando SweetViz en formato HTM con el código a continuación para revisar la distribución de los datos.

```python
# Abrir data frames
delitos_df = pd.read_parquet('https://github.com/osirisizs/Proyecto-Ingenieria-de-Caracteristicas/raw/main/delitos_df.parquet')
percepcion_df = pd.read_parquet('https://github.com/osirisizs/Proyecto-Ingenieria-de-Caracteristicas/raw/main/percepcion_df.parquet')

#%% Probando pandas-profiling

from pandas_profiling import ProfileReport 

profile = ProfileReport(
    delitos_df, 
    explorative=True,
    title='Reporte de delitos_df', 
    html={'style':{'full_width':True}}
) 
profile.to_file("delitos-pandas-profiler.html")

profile = ProfileReport(
    percepcion_df, 
    explorative=True,
    title='Reporte de percepcion_df', 
    html={'style':{'full_width':True}}
) 
profile.to_file("percepcion-pandas-profiler.html")
```
